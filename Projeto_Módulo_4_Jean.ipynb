Importei a biblioteca Pandas

```python
import pandas as pd
```

# **Dados de Junho**üìÖ

```python
junho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-06-gasolina-etanol.csv', sep=';')
junho.head()
```

```python
junho['Valor de Venda'] = junho['Valor de Venda'].apply(lambda x:x.replace(',', '.')).astype(float)#transformando o campo valor de venda pra float
del junho['Valor de Compra'] #excluindo a coluna vazia
```

# **Dados de JulhoüìÖ**


```python
julho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-07-gasolina-etanol.csv', sep=';')
julho
```

```python
julho['Valor de Venda'] = julho['Valor de Venda'].apply(lambda x:x.replace(',', '.')).astype(float)#transformando o campo valor de venda pra float
del julho['Valor de Compra'] #excluindo a coluna vazia
```

Exibi os dois links acima e depois os juntei em um √∫nico dataframe

```python
dataframe_JyJ = pd.concat([junho, julho],ignore_index=True)
dataframe_JyJ
```

# **3 - Quais os 5 estados com pre√ßo m√©dio de gasolina e etanol mais caros?**

```python
dataframe_JyJ.info()
```

```python
# dataframe_JyJ['Valor de Venda']=dataframe_JyJ['Valor de Venda'].str.replace(',','.') 
dataframe_JyJ['Valor de Venda']=dataframe_JyJ['Valor de Venda'].astype(float)
etanol = dataframe_JyJ[dataframe_JyJ['Produto']=='ETANOL'].groupby(['Estado - Sigla', 'Produto']).agg( {"Valor de Venda":"mean"}).round(2).sort_values('Valor de Venda',ascending=False).head(5)
gasolina= dataframe_JyJ[dataframe_JyJ['Produto'].str.contains('GASOLINA$')].groupby(['Estado - Sigla', 'Produto']).agg( {"Valor de Venda":"mean"}).round(2).sort_values('Valor de Venda',ascending=False).head(5)
# gasolina_adt= dataframe_JyJ[dataframe_JyJ['Produto'].str.contains('GASOLINA AD')].groupby(['Estado - Sigla', 'Produto']).agg( {"Valor de Venda":"mean"}).round(2).sort_values('Valor de Venda',ascending=False).head(5)

cinco_estados = pd.concat([etanol, gasolina], join="outer")
cinco_estados
```

#Conclu√≠mos que os cinco estados com o pre√ßo m√©dio de gasolina e etanol mais caros s√£o, para Etanol: 


1. **Rio Grande do Sul**
2. **Acre**
3. **Rio Grande do Norte**

#Para Gasolina, foram os seguintes estados:
