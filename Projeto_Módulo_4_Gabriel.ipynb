```python
#Import do Pandas e numpy
import pandas as pd
import numpy as np
```

# Análise de preços referentes aos combustíveis nos meses de Junho e Julho de 2021.

A análise em questão busca trazer informações que mostre em qual localidade do Brasil o maior e menor preço (gasolina, gasolina aditivada e etanol) são praticados.
Tem como base para análise os dados da ANP (Agência Nacional do Petróleo, Gás Natural e Biocombustíveis)

```python
#Função para exibição da tabela que será analisada usando o mês de Junho como referência.
junho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-06-gasolina-etanol.csv', sep=';')
junho
```

```python
#Função para exibição da tabela que será analisada usando o mês de Julho como referência.
julho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-07-gasolina-etanol.csv', sep=';')
julho
```

```python
#Função utilizada para a junção das tabelas de Junho e julho, transformando assim em uma única tabela.
df = pd.merge(junho, julho, how='outer')
df.head()
```

```python
#Função utilizada para criação de tabela com os produtos "Gasolina" e "Gasolina aditivado" para poder realizar a consulta por produto.
gasolina = ['GASOLINA', 'GASOLINA ADITIVADA']
selecao = df['Produto'].isin(gasolina)
tb_gasolina = df[selecao]
tb_gasolina
```

```python
#Função que apresenta o resultado do valor mínimo da gasolina na tabela em análise. 
tb_gasolina.min()
```

```python
#Função que apresenta o resultado do valor máximo da gasolina na tabela em análise.
tb_gasolina.max()
```

```python
#Função utilizada para criação de tabela com o produto "Etanol" para poder realizar a consulta por produto.
etanol = ['ETANOL']
selecao = df['Produto'].isin(etanol)
tb_etanol = df[selecao]
tb_etanol
```

```python
#Função que apresenta o resultado do valor mínimo do etanol na tabela em análise. 
tb_etanol.min()
```

```python
#Função que apresenta o resultado do valor máximo do etanol na tabela em análise. 
tb_etanol.max()
```
