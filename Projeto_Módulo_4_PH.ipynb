### Importando os arquivos csv referentes aos meses de JUNHO e JULHO (links acima) para depois juntá-los num único dataframe:

```python
junho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-06-gasolina-etanol.csv', sep=';')
junho.head()
```

```python
julho = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-07-gasolina-etanol.csv', sep=';')
julho.head()
```

### Juntando os Dataframes de Junho e Julho

```python
df = pd.merge(junho, julho, how='outer')
df.head()
```

### **LIMPANDO O DATAFRAME (REMOÇÃO DAS INFORMAÇÕES NÃO ÚTEIS À ANÁLISE):**

### *Checando se todos os valores de compra são nulos. Caso sejam, poderão ser removidos já que de nada servirão:*

```python
print(df['Valor de Compra'].unique())
```

### Checando se só há uma unidade de medida (R$/litro)

```python
print(df['Unidade de Medida'].unique())
```

### Removendo as colunas: CNPJ da Revenda, Nome da Rua, Número Rua, Complemento, Cep, Valor de Compra, Unidade de Medida

```python
df.drop(['CNPJ da Revenda', 'Nome da Rua', 'Número Rua','Complemento','Cep','Valor de Compra','Unidade de Medida'], axis=1, inplace=True)
```

```python
df.head() ##Exibindo colunas que ficaram 
```

### Para cálculos numéricos, é necessário que o valor de venda esteja como numérico. Se tiver como string, o cálculo não será feito. Checando se Valor de Venda é numérico ou não.

```python
df.info()
```

### Como o valor de venda é uma string (object), a vírgula será substituída por ponto e depois o valor será convertido de objeto para numérico (float) de forma a preparar o campo para o cálculo.

```python
df['Valor de Venda'] = df['Valor de Venda'].replace(',', '.', regex=True).astype(float)
df.head()
```

### Checando se converteu para float

```python
df.info()
```

### Com o Valor de Venda convertido para Float, será possível calcular o valor médio dos combustíveis por região. 

### **QUESTÃO 9: HÁ ALGUMA CORRELAÇÃO ENTRE O VALOR DO COMBUSTÍVEL (GASOLINA E ETANOL) E A REGIÃO ONDE ELE É VENDIDO?** ✅

### No caso abaixo, **a região SUDESTE é a que apresenta o menor valor médio dos combustíveis**, ficando o **2° lugar para a Região Centro Oeste**, **3° para a região Sul**, **4° para a Região NE** e o **N com o valor médio mais alto**.

```python
df.groupby('Região - Sigla').mean()
```

```python
df.groupby('Região - Sigla').mean().plot.bar(figsize=(12, 10))
```

### Analisando o **desvio Padrão do valor de Venda** dos combustíveis, é possível observar que a **região Centro Oeste é a que apresenta o maior valor**, **seguida da região Sudeste, Sul, Norte e Nordeste, respectivamente**. Por essa análise, onde se considera o preço de todos os combustíveis analisados (Gasolina Aditivada, Gasolina e Etanol), podemos observar que **a busca por preços mais baixos por parte do consumidor compensa mais na região Centro Oeste (maior variação) do que na região Nordeste (menor variação)**.

```python
df.groupby('Região - Sigla').std()
```

```python
df.groupby('Região - Sigla').std().plot.bar(figsize=(12, 10)) ## Plotando o gráfico de barras (plot.bar) e configurando o tamanho figsize=)
```

### Procedendo uma análise mais minuciosa onde se destaca o tipo de combustível, podemos observar que em relação ao preço do Etanol, **a região Sudeste apresenta o menor preço médio de venda, seguido do Centro Oeste.** Isso pode ser explicado pelo fato de grande parte do complexo sucroalcooleiro brasileiro estar concentrado nessas duas regiões (1° no Sudeste, sobretudo em São Paulo, e 2° no Centro Oeste que conta com Goiás e Mato Grosso). A proximidade desses complexos irá impactar nos custos de frete, uma vez que a maior parte do frete de etanol é feito via rodoviária (https://www.epe.gov.br/sites-pt/publicacoes-dados-abertos/publicacoes/PublicacoesArquivos/publicacao-615/NT-EPE-DPG-SDB-2021-03_Analise_de_Conjuntura_dos_Biocombustiveis_ano_2020.pdf) e impostos (As menores alíquotas de ICMS estão em parte do Sudeste e parte do Centro Oeste, onde podem ser metade do percentual de outras regiões com mínimo de 12% e Máximo de 30%). Apesar do Nordeste também possuir uma boa quantidade de usinas de Etanol, o ICMS do NE pressiona os valores para cima.
### Com relação à Gasolina (comum e aditivada) o padrão se "repete", com Sudeste com preços mais baixos provavelmente devido à localização de refinarias, menores custo de frete entre as refinarias e os postos de abastecimento e a maior presença da infraestrutura de distribuição por dutos, além da questão do ICMS já descrita acima.

```python
df.groupby(['Região - Sigla','Produto']).mean() ##Calculando a média de cada combustível e agrupando por região
```

```python
df.groupby(['Região - Sigla', 'Produto']).mean().plot.bar(figsize=(15, 10)) ## plotando o gráfico de barras para o cálculo acima
```

### Exibindo o resultado somente para o produto **GASOLINA por região**, podemos observar que a **região Centro Oeste possui o maior o maior preço médio para GASOLINA COMUM do Brasil**. Para o Produto **GASOLINA ADITIVADA**, a **Região Nordeste exibe o maior valor, seguido da região Centro Oeste**.

```python
df[df['Produto'].str.contains('GASOLINA')].groupby(['Região - Sigla', 'Produto']).mean() ## Filtrando os valores de gasolina (Comum e Aditivada) por região e calculando a média
```

```python
df[df['Produto'].str.contains('GASOLINA')].groupby(['Região - Sigla', 'Produto']).mean().plot.bar(figsize=(12, 10)) ##Plotando o gráfico de barras para o cálculo acima
```
